```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
### ZADANIE 6 (BONUS)

## Analiza korelacji cena-ocena

# Które produkty są "okazją" — ocena >= 4.5 i cena < 1000 zł? Wypisz nazwy i ceny.
# Jaka jest średnia cena produktów z oceną >= 4.0 vs < 4.0?
# Podpowiedź: łączenie warunków: (warunek1) & (warunek2)

przefiltrowane_ponizej = [x for x in oceny_klientow if x < 4.0]
przefiltrowane_powyzej = [x for x in oceny_klientow if x >=4.0]

for i in range(len(nazwy)):
    if oceny_klientow[i] >= 4.0:
      print("---PRODUKTY Z OKAZJA---")
      if ceny[i] < 1000:
         print("Nazwa produktu: ", nazwy[i])
         print("Cena: ", ceny[i])
   
srednia_powyzej = sum(przefiltrowane_powyzej) / len(przefiltrowane_powyzej)
print("Średnia cen produktów powyżej oceny 4.0: ", srednia_powyzej)
srednia_ponizej = sum(przefiltrowane_ponizej) / len(przefiltrowane_ponizej)
print("Średnia cen produktów ponizej oceny 4.0: ", srednia_ponizej)

    
       

        
```

```python
###ZADANIE 5
#Analiza zapasów


# Produkty z niskim stanem magazynowym (< 30 szt.) — nazwy
# Łączna wartość tych produktów
# Ile % całego magazynu stanowią?
wartosc_produktow         = 0
ilosc_czesciowy_produktow = 0

for i in range(len(nazwy)):
    if ilosci_magazyn[i] < 30:
        print("Nazwa produktu: ", nazwy[i])
        print("Cena produktu: ", ceny[i])
        print("Ilość w magazynie: ", ilosci_magazyn[i])

        ilosc_czesciowy_produktow =+ ilosci_magazyn[i]
        wartosc_produktow         =+ ceny[i]

suma_produktow = ilosci_magazyn.sum()
procent        = (ilosc_czesciowy_produktow / suma_produktow ) * 100       
print("Łączna wartość owych produktów: ", wartosc_produktow)
print("Procent produktów z całego magazynu: ", procent)
```
